```python
import torch
import numpy as np
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype = torch.float)
print(f"Random Tensor: \n {x_rand} \n")
```

```python
shape = (2,3,)
ones_tensor = torch.ones(shape)
rand_tensor = torch.rand(shape)
zero_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zero_tensor} \n")
```

```python
tensor = torch.rand(3,4)
print(tensor.shape) # Shape of Tensor
print(tensor.dtype) # Tensor Datatype
print(tensor.device) #Device that tensor is stored on.
```

```python
if torch.accelerator.is_available():
    tensor = tensor.to(torch.accelerator.current_accelerator())
```

```python
tensor = torch.ones(4,4)
print(tensor[0]) # first row
print(tensor[:,0]) # first column
print(tensor[..., -1]) # last column
tensor[:,1] = 0
print(tensor)
```

```python
# You can use torch.cat to concatenate a sequence of tensors along a given dimension.
t1 = torch.cat([tensor, tensor, tensor], dim = 1)
print(t1)
```

```python
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out = y3)
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```
